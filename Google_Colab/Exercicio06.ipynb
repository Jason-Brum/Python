```python
num = 6/7

outro_num = 10

t1 = "Hello"

t2 = "World"
 #tudo isso é comentário e não interfere no código

print('Resultado: {0:f}'.format(num))

print('Resultado: {0:.2f}'.format(num))

''' Este também é um comentário,
porém multilinhas.
É dessa maneira que
se faz um comentário
com mais de uma linha '''

print('Resultado: {0:8.2f}'.format(num))

print('Resultado: {0:8d}'.format(outro_num))

print('{0:10}{1:10}'.format(t1, t2))

print('{0:^10}{1:^10}'.format(t1, t2))

print('{0:>10}{1:>10}'.format(t1, t2))
```
